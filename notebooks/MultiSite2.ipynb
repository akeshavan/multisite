First, some imports and functions:

```python
import prettyplotlib as ppl
from mpl_toolkits.axes_grid.anchored_artists import AnchoredText
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std
np.random.seed(9876789)
from scipy.stats import zscore
%pylab inline
from utils import *
from copy import deepcopy
```

```python
def sim_data(n_hc,n_ms,age_hc_range,age_ms_range,
             alpha_a,alpha_d,b0,d0,sig_b,sig_d,
             sig_s,alpha,gamma):
    
    o = {}
    
    o['age_hc'] = linspace(*age_hc_range,num = n_hc)
    o['age_ms'] = linspace(*age_ms_range,num = n_ms)
    
    o['dis_1'] = np.random.normal(d0,sig_d[0],n_ms) 
    o['BV_hc_1'] = alpha_a*o['age_hc'] + np.random.normal(b0,sig_b[0],n_hc)
    o['BV_ms_1_real'] = alpha_a*o['age_ms'] + np.random.normal(b0,sig_b[0],n_ms) + alpha_d * o['dis_1'] + d0
    o['BV_ms_1'] = alpha[0] * o['BV_ms_1_real'] + gamma[1]
    
    o['dis_2'] = np.random.normal(d0,sig_d[0],n_ms) 
    o['BV_hc_2'] = alpha_a*o['age_hc'] + np.random.normal(b0,sig_b[0],n_hc)
    o['BV_ms_2_real'] = alpha_a*o['age_ms'] + np.random.normal(b0,sig_b[0],n_ms) + alpha_d * o['dis_2'] + d0
    o['BV_ms_2'] = alpha[0] * o['BV_ms_2_real'] + gamma[1]
    
    
    o['BV_all_ms'] = np.hstack((o['BV_ms_1'],o['BV_ms_2']))
    o['dis_all'] = np.hstack((o['dis_1'],o['dis_2']))
    o['age_all_ms'] = np.hstack((o['age_ms'],o['age_ms']))
    
    
    o['age_all_hc'] = np.hstack((o['age_hc'],['age_hc']))
    o['BV_all_hc'] = np.hstack((o['BV_hc_1'],o['BV_hc_2']))
    
    
    o['protocol_ms'] = np.hstack((np.ones(n_ms),np.zeros(n_ms)))
    
    return o
    
```

```python
def run_calib(o,plot=False,write=False):
    o['calib'] = run_model([o['BV_hc_1']],o['BV_hc_2'],sm.OLS,write)
    o['BV_ms_1_calib'] = o['calib'].params[1] * o['BV_ms_1'] + o['calib'].params[0]
    o['BV_all_calib'] = np.hstack((o['BV_ms_1_calib'],o['BV_ms_2']))
    if plot:
        fig, ax = subplots(1,2,figsize=(10,4))
        plot_corr([o['BV_hc_1']],o['BV_hc_2'],["$\hat{BV_1}$"],"$\hat{BV_2}$","Calibration",ax[0])
        plot_corr([o['BV_ms_1_calib']],o['BV_ms_2'],["$\hat{BV_{1,calib}}$"],'$\hat{BV_2}$',"Calibrated Data",ax[1])
    
```

```python
def run_separate_models(o,plot=False,write=False):
    
    if write: print "Site 1\n"
    o['s1_ms'] = run_model([o['age_ms'],o['dis_1']],o['BV_ms_1'],sm.RLM,write)
    if write: print "\nSite 2\n"
    o['s2_ms'] = run_model([o['age_ms'],o['dis_2']],o['BV_ms_2'],sm.RLM,write)
    
    if plot:
        plot_coef_split([o['s1_ms'].params,o['s2_ms'].params],
                [o['s1_ms'].bse,o['s2_ms'].bse],
                ["const","age","disease"],
                ["Site 1","Site 2"])
```

```python
def run_combined_models(o,plot=False,write=False):
    o['combine_v1_ms'] = run_model([o['age_all_ms'],o['dis_all']],o['BV_all_ms'],sm.RLM,write)
    o['combine_v2_ms'] = run_model([o['age_all_ms'],o['dis_all'],o['protocol_ms']],o['BV_all_ms'],sm.RLM,write)
    o['combine_v3'] = run_model([o['age_all_ms'],o['dis_all']],o['BV_all_calib'],sm.RLM, write)
    if plot:
        plot_coef_split([o['combine_v1_ms'].params, o['combine_v2_ms'].params[:3], o['combine_v3'].params],
          [o['combine_v1_ms'].bse, o['combine_v2_ms'].bse[:3], o['combine_v3'].bse],
          ["const","age","disease"],
          ["Combined","Protocol", "Calibrated"],cmap = ppl.set2[2:])
```

```python
def grid(o,n_range,parameter):
    coefs = np.zeros((3,len(n_range)))
    errs = np.zeros((3,len(n_range)))
    
    for i,n in enumerate(n_range):
        o[parameter] = n
        sim6 = sim_data(**o)
        run_calib(sim6)
        run_separate_models(sim6)
        run_combined_models(sim6)
        coef = sim6["combine_v3"].params
        err = sim6["combine_v3"].bse
        coefs[:,i] = coef
        errs[:,i] = err
        
    return coefs, errs
```

```python
def plot_grid(coefs,errs,hc_age_range,name):
    fig,ax = subplots(1,3,figsize=(20,4))
    ax[0].errorbar(hc_age_range,coeefs[0,:],marker='o',yerr=errs[0,:]);
    ax[1].errorbar(hc_age_range,coeefs[1,:],color=ppl.set2[1],marker='o',yerr=errs[1,:]);
    ax[2].errorbar(hc_age_range,coeefs[2,:],color=ppl.set2[2],marker='o',yerr=errs[2,:]);
    suptitle("Calibrated & Combined Coefficients",size=14)
    ax[0].set_xlabel(name)
    ax[1].set_xlabel(name)
    ax[2].set_xlabel(name)
    ax[0].set_ylabel("Intercept")
    ax[1].set_ylabel("Age")
    ax[2].set_ylabel("Disease")
```

### What happens to the combined,calibrated coefficients when we increase the number of controls?

```python
hc_age_range = range(10,100,5)
coeefs, errs = grid(p6,hc_age_range,"n_hc")
plot_grid(coeefs,errs,hc_age_range,"n_hc")
p6
```

**Conclusion**: Maybe it doesn't matter how many subjects are used for the calibration if the other parameters shown above are correct

### What happens to the combined, calibrated coefficients when we increase brain volume variation?

```python
p7 = deepcopy(p5)
sig_b_range = [np.array([0.001,0.001])*i for i in range(1,100,5)]
coeefs, errs = grid(p7,sig_b_range,"sig_b")
plot_grid(coeefs,errs,[x[0] for x in sig_b_range],"sig_b")
p7
```

**Conclusion**: If brain volumes vary by more than **0.03 Liters**, our calibration is pretty bad.

### What happens if we increase the weight of age on BV, while keep BV variation high (at 0.08), and with 10 controls?

```python
p7 = deepcopy(p5)
alpha_a_range = [i/1000. for i in range(5,20)]
coeefs, errs = grid(p7,alpha_a_range,"alpha_a")
plot_grid(coeefs,errs,alpha_a_range,"alpha_a")
p7
```

**Conclusion**: if the weight of age on Brain volume is high, then the disease and intercept coefficients are closer to what they should be.
