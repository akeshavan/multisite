##Background

* We want to simulate the MRI data we acquire from two different sites, for two groups (patients and controls) and test the  relationship with age and a disease metric (for patients only)

* We want to calculate a **linear calibration transformation** from site 1 to site 2 for the controls

* Our MRI metric will be brain volume (BV)

* We assume noise is normally distributed

* The control cohort will be subjects from the same population that are scanned at both sites. 
  * There are 50 controls
  
* There will be 250 patients from the same population at each site

* Within sites we want to test 2 models:
  1. OLS with age and brain volume for controls
  2. Robust Linear Model (RLM) with age, disease metric, and brain volume for patients

* We want to test 3 models to estimate the effect of EDSS on BV across sites:
  1. Combining the variables (age, disease) from both sites
  2. Combining the variables and adding a regressor for site (i.e. Protocol)
  3. Calibrating the measured brain volume from sites 1 and 2, then running the model
  
 
## Simulation Conclusions:

What conditions make the calibration superior?

First, assume a different number of patients at each site, with different distributions of EDSS.

* **Number of Patients (symmetric)**: Calibration is increasingly better as the number of patients at each site increases
* **Number of Controls**: Calibration is only better for 25 controls or greater. For less than 25, use protocol model
* **EDSS Distribution (asymmetric)**: Calibration gets better as the distribution becomes more asymmetric across sites for the site with more subjects. The protocol model is better if the site with less subjects has the asymmetric distribution.
* **Scanner Scaling coefficient**: As the difference in the two scaling coefficients increase, the calibration becomes increasingly beneficial. When there is a small difference between scaling, it doesn't matter which method to use.
* **Age range of patients (asymmetric)**: Has no effect on the model differences
* **Number of Patients (asymmetric)**: As asymmetry increases, calibration benefit increases slightly 

##Goal: Check Calibration Quality

We want to calculate the number of subjects needed to show ***equivalency*** of calibrated brain volume measurements between two different scanners. 

## Governing Equations

At each site, 50 age-matched controls from the same population are scanned. The scanner variability and subject variability are estimated from literature and from data in the lab. 

First, we define brain volume (BV) as a function of age for our healthy controls, with normally distributed variability:

###$BV = \alpha_a * A + N(B_0,\sigma_b)$

After being scanned at a site, the scanner scales the real brain volume and adds normally distributed scanner noise centered on an offset value:

At site 1:
###$\hat{BV_1} = \alpha_{s,1}*BV + N(\gamma_{s,1}, \sigma_{s,1})$

At site 2:
###$\hat{BV_2} = \alpha_{s,2}*BV + N(\gamma_{s,2}, \sigma_{s,2})$

##Equivalence Testing

The image below describes the equivalence test between two groups. The Null Hypothesis is that the two groups are different (like C and D in the diagram). The alternate hypothesis is that the differences between the two groups lie within a range of clinical/scientific indifference (like A and B). In the case ofE and F, there needs to be more subjects.

![equivalence image](http://web.archive.org/web/20120119090119im_/http://graphpad.com/library/graphics/equivalence.png)

We define this zone of indifference, or similarity margin, in terms of standard deviations of the brain volume differences, $\delta$. 

````python
_,ax = subplots(1)
pdiff = np.abs(z_euro_1-z_euro_2)/z_euro_2
sns.kdeplot(pdiff*100,shade=True,ax=ax)
title("Pairwise Difference in Calibrated BV between two sites")
ppl.remove_chartjunk(ax,["top","right"])
ax.set_xlim([0,5]);
ax.set_xlabel("Absolute pairwise difference (%)")
ax.set_ylabel("Frequency");
ax.vlines(np.mean(pdiff*100),0,.60);
at = get_label("$\mu$ = %2.2f\n$\sigma$ = %2.2f"%(np.mean(pdiff)*100,np.std(pdiff)*100),2)
ax.add_artist(at);
````

The graph above shows the spread of pairwise brain volume differences across sites. The mean is around 1% and the standard deviation is 0.7%. If we wanted to define our region of indifference as 1% change in BV, our equivalence margin, $\epsilon$ is 1/0.7 = 1.4 standard deviations. The function below runs the paired equivalence test in R, through python:

````python
def eq_test(pdiff,delta):
    x = ro.FloatVector(pdiff)
    base = ro.packages.importr("base")
    e = delta/std(pdiff)
    res = eq.ptte_data(x,Epsilon=e)
    P = res.rx("Power")[0][0]
    T = res.rx("Tstat")[0][0]
    C = res.rx("cutoff")[0][0]
    return P,T,C
````

Lets test similarity where we define an acceptible difference to be a 1% difference in brain volume:

````python
P,T,C = eq_test(pdiff,.01)
print "Power: ", P
print "Similar:", C > T
print C,T
````

With our 10 subjects, our power is high enough to detect similarity within a 1% BV difference. However, our calibration results in brain volumes that are statistically **NOT** similar! What if we change our margin to 2% ?

````python
P,T,C = eq_test(pdiff,.02)
print "Power: ", P
print "Similar:", C > T
print C,T
````

Well, now we see that our brain volumes **are statistically similar within 2%** after calibration. 

Next we calculate the power of our paired equivalence test by changing the equivalence margin:

````python
Ps = []
Ts = []
Cs = []
test_E = arange(0.001,0.01,0.0005)
for e in test_E:
    p,t,c = eq_test(pdiff,e)
    Ps.append(p)
    Cs.append(c)
    Ts.append(t)
````

````python
def eq_test(y,ep):
    import rpy2.robjects as ro
    eq = ro.packages.importr("equivalence")
    import numpy as np
    x = ro.FloatVector(y)
    e = ep/np.std(y)
    res = eq.ptte_data(x,Epsilon=e)
    P = res.rx("Power")[0][0]
    T = res.rx("Tstat")[0][0]
    C = res.rx("cutoff")[0][0]
    return P,T,C
````

````python
def test_all_parallel(pdiff,test_E = arange(0.001,0.01,0.0005),view=None):
    g = view.map(eq_test,[pdiff]*len(test_E),test_E)
    g = np.asarray(g)
    return g
````

## 4. HC Population Bias Detection

Now suppose there is a bias in the populations at the two sites. We expect to be able to detect the bias using equivalence testing. First, we introduce bias to our governing equiations:

###$BV_{1,b} = \alpha_{b,1}*BV_1 + \gamma_{b,1}$

Where $BV_1$ is the real brain volume:

###$BV_1 = \alpha_a*A + BV_0$

Then, when we scan,

###$\hat{BV_1} = \alpha_1*(\alpha_{b,1}*BV_1+\gamma_{b,1}) + \gamma_1 $

Substituting the real brain volume $BV_1$:

###$\hat{BV_1} = \alpha_1*(\alpha_{b,1}*(\alpha_a*A + BV_0)+\gamma_{b,1}) + \gamma_1 $

Therefor our measured age coefficient will be $\alpha_a*\alpha_{b,1}*\alpha_1$

````python
params["alpha_hc"] = [1, 1.1]
params["gamma_hc"] = [0, 1]
age_euro = np.array([51,32,24,30,32,37,39,24,24,37,29,32,40,41])
````

##Conclusions

* We need **at least 10 multisite controls**.
* We need **at least 50 healthy controls at each site**.
* Multisite controls need to be **scanned twice**.
* We can detect bias by choosing an equivalence margin of **3%** - meaning that if we pass the equivalence test, there is likely to be very little bias.
* Bias affects our EDSS coefficient estimate, but not badly. However, the power of the calibration estimate decreases in comparison to the power of the protocol estimate as bias increases. 
* The protocol estimate is very inaccurate because of the parameters chosen for this simulation (high difference between scaling factors.
