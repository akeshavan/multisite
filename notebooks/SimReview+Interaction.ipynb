##Background

* We want to simulate the MRI data we acquire from two different sites, for two groups (patients and controls) and test the  relationship with age and a disease metric (for patients only)

* We want to calculate a **linear calibration transformation** from site 1 to site 2 for the controls

* Our MRI metric will be brain volume (BV)

* We assume noise is normally distributed

* The control cohort will be subjects from the same population that are scanned at both sites. 
  * There are 50 controls
  
* There will be 250 patients from the same population at each site

* Within sites we want to test 2 models:
  1. OLS with age and brain volume for controls
  2. Robust Linear Model (RLM) with age, disease metric, and brain volume for patients

* We want to test 3 models to estimate the effect of EDSS on BV across sites:
  1. Combining the variables (age, disease) from both sites
  2. Combining the variables and adding a regressor for site (i.e. Protocol)
  3. Calibrating the measured brain volume from sites 1 and 2, then running the model
  

##Simulation Parameters

We will first define the following parameters, and then explain the origin of each one:

```python
params = {'n_hc':50, #number of controls
          'n_ms':[250,100], #number of MS patients at each site
          'age_hc_range':(20,52), # Age range of the controls
          'age_ms_range':[(20,45),(20,75)], # Age range of MS patients at each site
          'd_dist':[.65,.88], # Distribution weighting of EDSS scores for each site
          'alpha_a':-3, # Real coefficient of age
          'alpha_d':[-16,-16], # Real coefficient of EDSS score
          'alpha_i':[-1,-1], #TODO: Interaction term for age and EDSS (get real value)
          'b0':1644, #Brain Volume Intercept (in ccs)
          'sig_b':[57,57], # Brain variation across subjects (in ccs) for each site
          'sig_s':[16,16], # Variability in brain volume measurement across scanners for each site
          'alpha':[0.7,1.3], # Scaling of each scanner
          'gamma':[20,-20] # Offset of each scanner
          }
```

## Governing Equations
Let us define a linear relationship between age, brain volume, and disease metric, with **no variability or noise** to start:

For controls with no disease, the relationship is:

###$BV = \alpha_a * A + BV_0$

Where A is age.

The brain volume of a patient is a function of both age and disease metric:

###$BV = \alpha_a * A + BV_0 + \alpha_d * D$

Where D is the disease metric. For now we assume no interaction between the disease metric and age.

We defined true values for this relationship by setting $\alpha_a = -3$, $\alpha_d = -16$ for each site, and the intercept $BV_0 = 1644$:

We know there is variablity in brain volume between subjects, so our actual brain volume for controls is:
    
###$BV_{hc} = \alpha_a * A + N(BV_0,\sigma_b)$

Where $\sigma_b$ is the standard deviation in brain volume between subjects, defined as 57 and N represents a normal distribution centered at $BV_0$

Similarly for patients, the actual brain volume for patients is:

###$BV_{ms} = \alpha_a * A + N(BV_0,\sigma_b) + \alpha_d * D +\alpha_i*D*A$


Where D represents the disease metric: EDSS. From our data we see a double normal distribution of scores: 

###$D_{dist} = d_{dist}*N(2,1.1) + (1-d_{dist})*N(6,1.1)$

and $d_{dist}$ represents the probability weighting of the normal distribution centered at EDSS = 2. A sample distribution for the MS patients at two different sites looks like:


```python
sim = sim_data(**params)
sns.kdeplot(sim['dis_1'],shade=True,label="Site 1");
sns.kdeplot(sim['dis_2'],shade=True, label="Site 2");
title("EDSS Distribution at two different sites")
ylabel("Frequency")
legend()
xlabel("EDSS");
xlim([0,10]);
```

Now we must account for the scanner measurement. The scanner applies a linear transformation on the actual brain volume measurement:

###$$\hat{BV} = \alpha * BV + N(\gamma,\sigma_s)$$

* $\hat{BV}$ is the measure brain volume
* $\gamma$ is the measurement offset
* $\sigma_s$ is the variability of the scanner
* $\alpha$ is the scaling of the scanners

Let us look at the distributions of brain volumes for the MS patients at each site:

```python
sns.kdeplot(sim['BV_ms_1'],shade=True,label="Site 1");
sns.kdeplot(sim['BV_ms_2'],shade=True, label="Site 2");
legend()
title("Brain Volume Distributions at two sites")
ylabel("Frequency")
xlabel("Brain Volume (cc)");
```

Here we see that the scaling at site 1 and site 2 really separates the two distributions

##Calibration 

In our calibration we want to acheive two goals:

1. Remove age dependence: We assume age affects brain volumes at the same rate in both patients and controls
2. Correct for scanner scaling ($\alpha_{1,2}$) differences

After this, our two sets of calibrated data may be combined, and the effect of disease on brain volume can be solved. We start by subtracting the brain volume of controls (from the fit) from the brain volume of patients at the age of the patient.

###$BV_{z,1} = \frac{BV_1 - M_{hc,1}(A_{ms})}{\alpha_{s,1}*\alpha_{a}}$

Where $M_{hc,1}$ is the fitted model of $BV_{hc,1}$ ~ $A_{hc}$ :

###$BV_{z,1} = \frac{BV_1 - (\alpha_{s,1}*(\alpha_a*A_{ms}+BV_0) + \gamma)}{\alpha_{s,1}*\alpha_{a}}$

Expanding the BV formula for MS patients:

###$BV_{z,1} = \frac{\alpha_{s,1}(\alpha_a*A_{ms}+\alpha_d*D +\alpha_i*A*D +BV_0)+\gamma - (\alpha_{s,1}*(\alpha_a*A_{ms}+BV_0) + \gamma)}{\alpha_{s,1}*\alpha_{a}}$

Reducing the equation gives:

###$BV_{z,1} = \frac{\alpha_{s,1}(\alpha_d*D + \alpha_i*A*D)}{\alpha_{s,1}*\alpha_{a}} = \frac{\alpha_d}{\alpha_a}*D + \frac{\alpha_i}{\alpha_a}*A*D$

First lets see what happens when we run the model many times, choosing a new set of subjects from the distribution each time. We will see the true confidence intervals of the coefficients this way. We will also iterate over the number of subjects in the MS group and see how this affects our power.

Not the asymmetry in the parameters between the two sites. There is a smaller number of ms subjects at the second site, along with a larger range of age distributions, and overall lower EDSS scores. Finally, the scaling of the scanner is not even between the two sites.

```python
def get_table(a,bold=None):
    data = [["\textParameter","Site 1","Site 2"]]
    for key, val in sorted(a.iteritems()):
        if isinstance(val,list):
            data.append([key, "%s"%str(val[0]), "%s"%str(val[1])])
        else:
            data.append([key,str(val),str(val)])
    txt = """\\begin{tabular}{llll}"""
    txtend = """\\end{tabular}"""
    for i,row in enumerate(data):
        txt+=" & ".join(row).replace("_"," ")
        txt+=("""\\\ \\hline """)
    txt += txtend
    return txt
```

```python
n_ms_range = [[50,50],[100,100],[250,250],[500,500]]
a = deepcopy(params)
cz,ez,iz,zz, cp,ep,ip,zp,ccc,ecc,icc,zcc = grid(a,n_ms_range,'n_ms',100)
#a["n_ms"] = ["??","??"]

foo = get_table(a)
```

We are mostly interested in the disease coefficient. First we look at the Protocol model. From our parameter definition we expect the coefficient to be $\alpha_a = -16$:

```python
#a.pop("n_ms")
foo = get_table(a)
fig,ax = subplots(1,3,figsize=(15,5))
ax1 = plot_grid_box(ccc[[2],:,:],[x[0] for x in n_ms_range],
                      ["Disease Coeffient"],
                      "Protocol Model","Number of MS Patients at each site",ax=ax[0])
ax1 = plot_grid_box(zcc[[2],:,:],[x[0] for x in n_ms_range],
                      ["Disease z score of coefficient"],
                      "Protocol Model","Number of MS Patients at each site",ax=ax[1])

ax[-1].text(0,0.1,foo,size=16);
ax[-1].grid("off");
ax[-1].axis("off");
```

Lets compare this result to running the calibrated model. From our parameter definition, we expect the disease coefficient to be $\alpha_d/\alpha_a = 16/3 = 5.33$

```python
fig,ax = subplots(1,3,figsize=(15,5))
ax0 = plot_grid_box(cz[[1],:,:],[x[0] for x in n_ms_range],
                      ["Disease Coeffient"],
                      "Protocol Model","Number of MS Patients at each site",ax=ax[0])
ax1 = plot_grid_box(zz[[1],:,:],[x[0] for x in n_ms_range],
                      ["Disease z score of coefficient"],
                      "Calibrated Model","Number of MS Patients at each site",ax=ax[1])
ax0[0].set_ylim([0,15]);
ax[-1].text(0,0.1,foo,size=16);
ax[-1].grid("off");
ax[-1].axis("off");
```

The see which model is better, we pair-wise subtact the coefficient z scores.

```python
fig,ax = subplots(1,2,figsize=(10,5))
diff_dis = zz[-1,:,:] + zcc[2,:,:]
foo = np.zeros((1,4,100))
foo[0,:,:] = diff_dis
ax0 = plot_grid_box(foo,[d[0] for d in n_ms_range],["$z_{calibration} - z_{protocol}$"],
                   "Difference Between Models ($\Delta$z)","Number of MS Patients at each site",ax=ax[0]);
ax[-1].text(0,0.1,get_table(a),size=16);
ax[-1].grid("off");
ax[-1].axis("off");
```

The clear winner is the calibrated model, because the difference in z scores for the coefficient estimates is positive. More importantly, the coefficient estimates are more accurate with the calibrated model, especially for a small number of subjects.

## Conclusions

What conditions make the calibration superior?

First, assume a different number of patients at each site, with different distributions of EDSS.

* **Number of Patients (symmetric)**: Calibration is increasingly better as the number of patients at each site increases
* **Number of Controls**: Calibration is only better for 25 controls or greater. For less than 25, use protocol model
* **EDSS Distribution (asymmetric)**: Calibration gets better as the distribution becomes more asymmetric across sites for the site with more subjects. The protocol model is better if the site with less subjects has the asymmetric distribution.
* **Scanner Scaling coefficient**: As the difference in the two scaling coefficients increase, the calibration becomes increasingly beneficial. When there is a small difference between scaling, it doesn't matter which method to use.
* **Age range of patients (asymmetric)**: Has no effect on the model differences
* **Number of Patients (asymmetric)**: As asymmetry increases, calibration benefit increases slightly
