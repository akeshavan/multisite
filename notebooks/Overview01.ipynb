### Freesurfer

All T1 scans were run through Freesurfer 5.3.0, with no manual correction with the exception of two manual talairach transform adjustments.

```python
df_v = pd.read_csv("../stats/aseg_fs_volume.csv",index_col=0)
```

Here, we calculate the within-site and between-site metrics:

S = $\frac{\sigma_m}{\mu_m} * 100$ = percent deviation from the mean

```python
metrics = df_v.columns[13:]
data = np.zeros((2,len(metrics)))
for i,m in enumerate(metrics):
    data[:,i] = get_variabilities(df_v,m)
```

Next, we calculate the difference in within-site and across-site metrics.

```python
vardf = pd.DataFrame({"within-site":data[0,:],"across-site":data[1,:]},index=metrics)
vardf["diff"] = vardf["across-site"] - vardf["within-site"]
vardf = vardf.sort("diff")
```

#### Basel Prisma MPRAGE vs. Milan Intera SENSE

```python
key = "Subcortical gray matter volume"
subcort_df = pd.pivot_table(df_v,cols="site-scanner",rows="subject_id",values = key)
sns.jointplot("ams-discovery-fspgr","mil-intera-sense",subcort_df,kind="reg")
model = sm.OLS(subcort_df["ams-discovery-fspgr"],subcort_df["mil-intera-sense"],missing='drop')
res = model.fit()
res.summary()
```

## Conclusions and Next Steps

* Because of FOV differences between sites, we can only get accurate measures of subcortical and temporal morphometry from **all** sites. 
* As such, the subcortical, corpus callosum, and hippocampal volumes had the lowest across site % deviation from the mean. 
* Total white matter volume had the largest discrepancy between across-site deviation(~20%) and within-site deviation(~0.75%). This can probably be attributed to the FOV differences across sites. Additionally, the SENSE sequence from Milan's Phillips Intera did not correlate well with measures from other sites. This could be because of contrast differences between the SENSE and MPRAGEs. This needs to be fixed in order to incorporate retrospective data in the analysis.

* The first thing freesurfer and FIRST do is calculate a transform of their MNI brain to the subject brain, but often this fails, especially for T1se and the SENSE scans, because the contrast is so different. 
* I think we can get a better segmentation from from probability-atlas based segmentation. The probability atlas will be built in the following steps:

  1. Create overall subject templates that include all scans across scanners for one subject
  2. Warp all freesurfer segments to each subject's template, and average to create a subject's probability atlas.
  3. Create sequence specific templates 
  4. Warp subject-probability maps to each subject's sequence, and then to the sequence-template. Average for all subjects and create a template probability atlas
  5. Warp the template probability atlas to each subject/scan's native space. 
  6. Run ANTS-Atropos segmentation and ANTS-Cortical Thickness pipelines which are based on the template 
  
* The major benefit of this method is that scans with bad segmentation from freesurfer can be corrected by the good segmentation scans via the probability atlas. My hope is that the across-site deviations wil decrease closer to the within-site variation for all metrics with this method.
