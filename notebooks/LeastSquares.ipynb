```python
import prettyplotlib as ppl
from mpl_toolkits.axes_grid.anchored_artists import AnchoredText
```

```python
def get_label(txt,loc=2):
    at = AnchoredText(txt,
                      prop=dict(size=12), frameon=True,
                      loc=loc,
                      )
    at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
    return at
```

### At one site:

$BV = \alpha_0 * age+B_0+N(0,\sigma_i^2)$

Where:

$\sigma_i$ = variance between subjects

```python
age = linspace(20,52,50)
alpha0 = 5
b0 = 2 
sig_i = 5
```

```python
fig, ax = subplots(1,1)
x = np.random.normal(0,sig_i,500)
ax.set_title("Normal Error Distrubution\n$\sigma_i = %s$"%sig_i)
ppl.hist(ax,x,bins=20);
```

```python
BV = alpha0*age + b0+ np.random.normal(0,sig_i,len(age))

fig,ax = subplots(1)
ax.set_xlabel("Age")
ax.set_ylabel("Brain Volume")
ppl.plot(ax,age,BV,marker='o',linewidth=0);
coef = np.polyfit(age,BV,1)
at = get_label("BV = %2.2f * age + %2.2f"%(coef[0],coef[1]))
ax.add_artist(at)
ppl.plot(ax,age,[coef[0]*a+coef[1] for a in age],linewidth=5);
```

### Now we have two sites with different subjects at each site:

$\hat{BV_1} = BV_{1} + N(0,\sigma_{s,1}^2)$

$\hat{BV_2} = BV_{2} + N(0,\sigma_{s,2}^2)$

Where

* $\sigma_{s,1}$ is the variance at site 1 

* $\sigma_{s,2}$ is the variance at site 2 

* $BV_1 = \alpha_0 * age+B_0+N(0,\sigma_{i,1}^2)$

* $BV_2 = \alpha_0 * age+B_0+N(0,\sigma_{i,2}^2)$

```python
sig_i = [.1,.1]#sorted(np.random.rand(2)*10)[::-1]
sig_s = sorted(np.random.rand(2)*1)[::-1]
```

```python
fig, ax = subplots(1,2,figsize=(12,4))
for i,sig in enumerate(sig_i):
    x = np.random.normal(0,sig,500)
    ppl.hist(ax[0],x,bins=20,label="$\sigma_{i,%d} = %2.2f$"%(i+1,sig));
    ax[0].set_title("Subject Variability")
ppl.legend(ax[0]);
    
for i,sig in enumerate(sig_s):
    x = np.random.normal(0,sig,500)
    ppl.hist(ax[1],x,bins=20,label="$\sigma_{s,%d} = %2.2f$"%(i+1,sig));
    ax[1].set_title("Site Variability")
ppl.legend(ax[1]);

    
#suptitle("Normal Error Distrubution");
```

```python
fig,ax = subplots(1,1)
ax.set_xlabel("Age")
ax.set_ylabel("Brain Volume")
txt=[]
BV_all = []
for i,sig in enumerate(sig_i):
    BV = alpha0*age + b0 + np.random.normal(0,sig,len(age))+np.random.normal(0,sig_s[i])
    BV_all.append(BV)
    ppl.plot(ax,age,BV,marker='o',label="Site %d"%(i+1),linewidth=0);
    coef = np.polyfit(age,BV,1)
    ppl.plot(ax,age,[coef[0]*a+coef[1] for a in age],linewidth=2,color=ppl.set2[i]);
    txt.append("$\hat{BV_%d} = %2.2f * age + %2.2f$"%(i+1,coef[0], coef[1]))
    
at = get_label('\n'.join(txt),4)
ax.add_artist(at)
    
ppl.legend(ax,loc=2);
```

```python
fig,ax = subplots(1,1)
ax.set_xlabel("$\hat{BV_1}$")
ax.set_ylabel("$\hat{BV_2}$")
ppl.plot(ax,BV_all[0],BV_all[1],linewidth=0, marker='o',color=ppl.set1[0])
coef = np.polyfit(BV_all[0],BV_all[1],1)
ppl.plot(ax, BV_all[0],[coef[0]*bv + coef[1] for bv in BV_all[0]],linewidth=2,color=ppl.set1[1])
at = get_label("$\hat{BV_2} = %2.2f * \hat{BV_1} + %2.2f$"%(coef[0],coef[1]),2)
ax.add_artist(at)
ax.set_title("Calibration");
```

### Add more variables and model different scanners

```python
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
from statsmodels.sandbox.regression.predstd import wls_prediction_std

np.random.seed(9876789)
```

Define some "actual" variables

```python
n_hc = 10
n_ms = 250
age_hc = linspace(20,52,n_hc)
age_ms = linspace(20,52,n_ms)
# add a new variable, disease, that doesn't correlate at all with age
#dis = sin(age)

#weights on the variables
alpha_a = -4./1000 # age weight
alpha_d = -.5 #disease weight

# Intercept
b0 = 1.5

d0 = 1
```

Noise variance for each measurement at each site

```python
#Brain Volume
sig_b = [0.05,0.05]

#Disease
sig_d = [0.2,0.2]

#Site
sig_s = [0.02,0.025] 

#MRI
alpha = [0.9,1.1]
gamma = [0.05,-0.05]
```

Run OLS for each site

```python
pred = []
BV_all = []
params = []
err = []
BV_hc_all =[]
for i,sig in enumerate(sig_b):
    
    dis = np.random.normal(d0,sig_d[i],n_ms) 
    #true BV
    BV_healthy = alpha_a*age_hc + np.random.normal(b0,sig_b[i],n_hc) #genetics
    BV_patient = alpha_a*age_ms + np.random.normal(b0,sig_b[i],n_ms) #genetics
    
    BV_patient = BV_patient+ alpha_d*dis
    
    # brain volume measurement:
    BV_p = alpha[i]*BV_patient +  np.random.normal(gamma[i],sig_s[i],n_ms)
    BV_h = alpha[i]*BV_healthy  + np.random.normal(gamma[i],sig_s[i],n_hc)
    
    #BV_all.append(BV)
    
    #controls models
    X_hc = age_hc
    X_hc = sm.add_constant(X_hc)
    model_hc = sm.OLS(BV_h,X_hc)
    res_hc = model_hc.fit()
    #print res_hc.summary()
    BV_hc_all.append(BV_h)
    ##params_hc.append(res_hc.params)
    #err_hc.append(res_hc.bse)
    #pred_hc.append(res_hc.predict())
    
    # Patients
    X_ms = np.column_stack((age_ms,dis))
    X_ms = sm.add_constant(X_ms)
    model_ms = sm.RLM(BV_p,X_ms)
    res_ms = model_ms.fit()
    print res_ms.summary()
    
    
```

```python
fig,ax = subplots(1,1)
ax.set_xlabel("$\hat{BV_1}$")
ax.set_ylabel("$\hat{BV_2}$")
ppl.plot(ax,BV_hc_all[0],BV_hc_all[1],linewidth=0, marker='o',color=ppl.set1[0])


X_calib = sm.add_constant(BV_hc_all[0])
model_ms = sm.RLM(BV_hc_all[1],X_calib)
res_calib = model_ms.fit()
print res_calib.summary()

#coef = np.polyfit(,pred[1],1)
#ppl.plot(ax, pred[0],[coef[0]*bv + coef[1] for bv in pred[0]],linewidth=2,color=ppl.set1[1])
#at = get_label("$\hat{BV_2} = %2.2f * \hat{BV_1} + %2.2f$"%(coef[0],coef[1]),2)
#ax.add_artist(at)
#ax.set_title("Calibration");
```
