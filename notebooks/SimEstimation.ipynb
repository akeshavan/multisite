```python
def sim_data(n_hc,n_ms,age_hc_range,age_ms_range,
             alpha_a,alpha_d,b0,sig_b,
             sig_s,alpha,gamma):
    
    o = {}
    
    o['age_hc'] = linspace(*age_hc_range,num = n_hc)
    o['age_ms'] = linspace(*age_ms_range,num = n_ms)
    
    o['dis_1'] = np.random.binomial(1,0.88,n_ms)*np.random.normal(2,1.1,n_ms) + np.random.binomial(1,0.12,n_ms)*np.random.normal(6,0.8,n_ms) 
    o['dis_1'][o['dis_1']<0] = 0 #EDSS Cannot be below 0
    o['dis_1'][o['dis_1']>10] = 10 #EDSS Cannot be >10
    
    o['BV_hc_1_real'] = alpha_a*o['age_hc'] + np.random.normal(b0,sig_b[0],n_hc)
    o['BV_hc_1'] = alpha[0] * o['BV_hc_1_real'] + gamma[0] + np.random.normal(0,sig_s[0],n_hc)
    
    o['BV_ms_1_real'] = alpha_a*o['age_ms'] + np.random.normal(b0,sig_b[0],n_ms) + alpha_d * o['dis_1'] 
    o['BV_ms_1'] = alpha[0] * o['BV_ms_1_real'] + gamma[0] + np.random.normal(0,sig_s[0],n_ms)
    
    o['dis_2'] = np.random.binomial(1,0.88,n_ms)*np.random.normal(2,1.1,n_ms) + np.random.binomial(1,0.12,n_ms)*np.random.normal(6,0.8,n_ms)
    o['dis_2'][o['dis_2']<0] = 0 #EDSS Cannot be below 0
    o['dis_2'][o['dis_2']>10] = 10 #EDSS Cannot be >10
    
    #o['BV_hc_2_real'] = same subjects, so this is BV_hc_1_real
    o['BV_hc_2'] = alpha[1]* o["BV_hc_1_real"] +  gamma[1] + np.random.normal(0,sig_s[1],n_hc)
    
    o['BV_ms_2_real'] = alpha_a*o['age_ms'] + np.random.normal(b0,sig_b[1],n_ms) + alpha_d * o['dis_2'] 
    o['BV_ms_2'] = alpha[1] * o['BV_ms_2_real'] + gamma[1] + np.random.normal(0,sig_s[1],n_ms)
    
    o['BV_all_ms'] = np.hstack((o['BV_ms_1'],o['BV_ms_2']))
    o['dis_all'] = np.hstack((o['dis_1'],o['dis_2']))
    o['age_all_ms'] = np.hstack((o['age_ms'],o['age_ms']))
    
    o['age_all_hc'] = np.hstack((o['age_hc'],['age_hc']))
    o['BV_all_hc'] = np.hstack((o['BV_hc_1'],o['BV_hc_2']))
    
    o['protocol_ms'] = np.hstack((np.ones(n_ms),np.zeros(n_ms)))
    
    return o
```

#### What if we run this 100 times, what is the average intercept,slope, and $R^2$ value?

```python
def run_hc_model(o):
    o["hc_model_real"] = run_model([o["age_hc"]],o["BV_hc_1_real"],write=False)
```

```python
def grid(o,n_range,parameter,n_boot=10):
    # These are for the calibration
    coefs = np.zeros((2,len(n_range),n_boot))
    errs = np.zeros((2,len(n_range),n_boot))
    
    #These are for the healthy controls
    real_coefs = np.zeros((2,len(n_range),n_boot))
    real_errs = np.zeros((2,len(n_range),n_boot))
    
    #site 1 and 2 models
    coefs_1 = np.zeros((2,len(n_range),n_boot))
    errs_1 = np.zeros((2,len(n_range),n_boot))
    coefs_2 = np.zeros((2,len(n_range),n_boot))
    errs_2 = np.zeros((2,len(n_range),n_boot))
    
    hc_calib_coefs = np.zeros((2,len(n_range),n_boot))
    hc_calib_errs = np.zeros((2,len(n_range),n_boot))
    
    for i,n in enumerate(n_range):
        for j in range(n_boot):
            o[parameter] = n
            sim6 = sim_data(**o)
            run_calib(sim6)
            run_hc_model(sim6)
            run_separate_models(sim6)
            run_combined_models(sim6)
            coef = sim6["calib"].params
            err = sim6["calib"].bse
            coefs[:,i,j] = coef
            errs[:,i,j] = err
            
            coef_real = sim6["hc_model_real"].params
            err_real = sim6["hc_model_real"].bse
            real_coefs[:,i,j] = coef_real
            real_errs[:,i,j] = err_real
            
            coefs_1[:,i,j] = sim6['hc_1_calib'].params 
            errs_1[:,i,j] = sim6['hc_1_calib'].bse
            
            coefs_2[:,i,j] = sim6['hc_2_calib'].params 
            errs_2[:,i,j] = sim6['hc_2_calib'].bse
        
            hc_calib_coefs[:,i,j] = sim6["hc_12_calib"].params
            hc_calib_errs[:,i,j] = sim6["hc_12_calib"].bse
        
        
    return coefs, errs, real_coefs, real_errs, coefs_1, errs_1,coefs_2, errs_2,hc_calib_coefs, hc_calib_errs
```

```python
def plot_grid(coefs,errs,hc_age_range,name,title="Calibration Coefficients"):
    fig,ax = subplots(1,2,figsize=(10,4))
    
    for j in range(coefs.shape[-1]):
        ax[0].errorbar(hc_age_range,coefs[0,:,j],color=ppl.set1[0],marker='o',yerr=errs[0,:,j]);
        ax[1].errorbar(hc_age_range,coefs[1,:,j],color=ppl.set1[1],marker='o',yerr=errs[1,:,j]);
    
    #ax[2].errorbar(hc_age_range,coeefs[2,:],color=ppl.set2[2],marker='o',yerr=errs[2,:]);
    suptitle(title,size=14)
    ax[0].set_xlabel(name)
    ax[1].set_xlabel(name)
    #ax[2].set_xlabel(name)
    ax[0].set_ylabel("Intercept")
    ax[1].set_ylabel("Age")
    ax[0].set_xlim([0.5,max(hc_age_range)+0.5])
    ax[1].set_xlim([0.5,max(hc_age_range)+0.5])
    #ax[2].set_ylabel("Disease")
```

## Checking Calibration

In the ideal case, the calibration slope should be 1 and the intercept should be 0 if we have many subjects and the scanner offests and slopes are identical.

Below is the calibration function:

```python
def run_calib(o,plot=False,write=False):
    o['calib'] = run_model([o['BV_hc_1']],o['BV_hc_2'],sm.OLS,write)
    o['BV_ms_1_calib'] = o['calib'].params[1] * o['BV_ms_1'] + o['calib'].params[0]
    o['BV_all_calib'] = np.hstack((o['BV_ms_1_calib'],o['BV_ms_2']))
    if plot:
        fig, ax = subplots(1,2,figsize=(10,4))
        plot_corr([o['BV_hc_1']],o['BV_hc_2'],["$\hat{BV_1}$"],"$\hat{BV_2}$","Calibration",ax[0])
        plot_corr([o['BV_ms_1_calib']],o['BV_ms_2'],["$\hat{BV_{1,calib}}$"],'$\hat{BV_2}$',"Calibrated Data",ax[1])
```

I've increased the number of controls to 100, and $\alpha$ and $\gamma$ are the same for both sites
