### Background

Given a .CSV file with the fields:

* age
* BV (Brain Volume)
* subject_type (MS or HC)
* edss (edss score, 0-10)
* edss_group (binary 1 = low, 0 = high)
* protocol (site names - UCSF, Milan, etc)
* subject (subject id)

1. Calibrate the data across sites in two ways:
  * age matched (BV_age_matched)
  * percentile (BV_percentile)
  
2. Check the calibration using multisite controls via equivalence testing given an equivalence margin. Return .CSV with calibration stats.
  * Columns: Site 1, Site 2, n_subs, mean_diff, std_diff, eq_margin, t_score, cutoff, Pass/NoPass
  * For all pairs of sites

### Age-Matched Calibration

```python
def run_model(xi,y, model=sm.OLS,write=True,add_intercept=True,yname=None,xnames=None,title=None):
    k = len(xi)
    X = xi[0]
    for i in range(1,k):
        X = np.column_stack((X,xi[i]))
    if add_intercept:
        X = sm.add_constant(X)
    m = model(y,X)
    if model == sm.Logit:
        res = m.fit(disp=False)
    else:
        res = m.fit()
    if write:
        print res.summary(yname=yname,xname=xnames,title=title)
    return res
```

Make it a function!

```python
def age_matched_ms_calibration(filename):
    df = pd.read_csv(filename,index_col=0)
    df["BV_age_matched"] = np.ones(len(df))*np.nan
    protocols = np.unique(df["protocol"])
    for protocol in protocols:
        hc_idx = df.age[df.protocol==protocol][df.subject_type == "HC"].index
        ms_idx = df.age[df.protocol==protocol][df.subject_type == "MS"].index
        hc_model = run_model([df.age[hc_idx]],df.BV[hc_idx],sm.OLS,False)
        age_matching = df.BV[ms_idx] - hc_model.predict(sm.add_constant(df.age[ms_idx]))
        df.BV_age_matched[ms_idx] = age_matching/hc_model.params[1]
    output = write_csv(df)
    return output, df
```

### Transformation Calibration

Transform data from other sites in terms of the first site. Run 2 models, and transform to the first model:

$BV_1 = \alpha_1 * age + \gamma_1$

$BV_2 = \alpha_2 * age + \gamma_2$

```python
def transformation_hc_calibration(filename):
    """
    Transforms HC BV only to first (alphabetical) protocol
    """
    df = pd.read_csv(filename,index_col=0)
    protocols = np.unique(df["protocol"]).tolist()
    first_protocol = protocols[0]
    
    df["BV_%s_calibrated"%first_protocol] = np.ones(len(df))*np.nan
    hc_idx = df.age[df.protocol==first_protocol][df.subject_type == "HC"].index
    first_protocol_model = run_model([df.age[hc_idx]],df.BV[hc_idx],sm.OLS,False)
    
    df["BV_%s_calibrated"%first_protocol][hc_idx] = df.BV[hc_idx]
    
    for protocol in protocols[1:]:
        hc_idx = df.age[df.protocol==protocol][df.subject_type == "HC"].index
        ms_idx = df.age[df.protocol==protocol][df.subject_type == "MS"].index
        
        hc_model = run_model([df.age[hc_idx]],df.BV[hc_idx],sm.OLS,False)
        
        site_calib = run_model([hc_model.fittedvalues],
                               first_protocol_model.predict(sm.add_constant(df.age[hc_idx])),
                               sm.OLS, True)
        #run_model([hc_model.fittedvalues],first_protocol_model.fittedvalues,sm.OLS,True)
        
        df["BV_%s_calibrated"%first_protocol][hc_idx] = site_calib.predict(sm.add_constant(df.BV[hc_idx]))
        
    output = write_csv(df)
    return output, df
```

### Percentile Calibration

```python
def to_percentile(sim,ref_key,key):
    foo = ECDF(sim[ref_key],side="right")
    return foo(sim[key])
```

```python
def percentile_calibration(filename):
    df = pd.read_csv(filename,index_col=0)
    protocols = np.unique(df["protocol"]).tolist()
    df["BV_percentile"] = np.ones(len(df))*np.nan
    
    for protocol in protocols:
        hc_idx = df.age[df.protocol==protocol][df.subject_type == "HC"].index
        ms_idx = df.age[df.protocol==protocol][df.subject_type == "MS"].index
        percentile = ECDF(df.BV[hc_idx],side="right")
        df.BV_percentile[hc_idx] = percentile(df.BV[hc_idx])
        df.BV_percentile[ms_idx] = percentile(df.BV[ms_idx])
        
    output = write_csv(df)
    return output, df
```

```python
def take_mean(data):
    if len(data) > 1:
        return np.mean(data).tolist()[0]
    else:
        return data.tolist()[0]
```

```python
def calculate_differences(subject,protocols,key,df,calib):
    s1, s2 = np.triu_indices(len(protocols),1)
    for i in range(len(s1)):
        p1 = protocols[s1[i]]
        p2 = protocols[s2[i]]
        name= "%s-%s"%(p1,p2)
        data1 = take_mean(df[key][df.protocol==p1][df.subject==subject])
        data2 = take_mean(df[key][df.protocol==p2][df.subject==subject])
        calib[name][calib.subject==subject] = np.abs(data1-data2)/np.mean([data1,data2])
    return calib
```

Put it all together and what do you get:

```python
def check_calibration(df,calibration_method):
    import os
    import numpy as np
    import pandas as pd
    #Multisite Controls
    msc = np.unique(df.subject[df.duplicated("subject")][df.subject_type=="HC"])
    protocols = np.unique(df.protocol).tolist()
    
    # Initialize a calibration data frame
    calib = pd.DataFrame(data={"subject":msc.tolist()})
    n = len(msc)
    for i in range(len(s1)):
        p1 = protocols[s1[i]]
        p2 = protocols[s2[i]]
        calib["%s-%s"%(p1,p2)] = np.ones(n)*np.nan
    
    for subject in msc:
        calculate_differences(subject,protocols,calibration_method,df,calib)
        
    outfile = os.path.abspath(calibration_method+".csv")
    calib.to_csv(outfile)
    return outfile, calib
```
