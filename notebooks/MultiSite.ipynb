# Multi-Site MRI Simulation

###Background

* We want to simulate the MRI data we acquire from two different sites, for two groups (patients and controls) and test the  relationship with age and a disease metric (for patients only)

* We want to calculate a **linear calibration transformation** from site 1 to site 2

* Our MRI metric will be brain volume (BV)

* We assume noise is normally distributed

* The control cohort will be subjects from the same population that are scanned at both sites. 
  * There are 10 controls
  
* There will be 250 patients from the same population at each site

* Within sites we want to test 2 models:
  1. OLS with age and brain volume for controls
  2. Robust Linear Model (RLM) with age, disease metric, and brain volume for patients

* We want to test 3 models to estimate BV across sites:
  1. Combining the variables (age, disease) from both sites
  2. Combining the variables and adding a regressor for site
  3. Calibrating the measured brain volume from site 2 and combining with site 1, then running the model with age and disease

### Simulated Data

Define number of healthy controls and ms patients. There are 10 controls and 250 patients at each site

```python
n_hc = 10
n_ms = 250
```

Define a range of ages for the control and patient subjects, between 20 and 52 years

```python
age_hc = linspace(20,52,n_hc)
age_ms = linspace(20,52,n_ms)
```

Let us define a linear relationship between age, brain volume, and disease metric, with **no variability or noise** to start:

For controls with no disease, the relationship is:

$BV = \alpha_a * A + BV_0$

Where A is age.

The brain volume of a patient is a function of both age and disease metric:

$BV = \alpha_a * A + BV_0 + \alpha_d * D + D_0$

Where D is the disease metric.

We must define true values for this relationship by setting $\alpha_a$, $\alpha_d$, and the intercepts $BV_0$ and $D_0$

```python
#weights on the variables
alpha_a = -5./1000
alpha_d = -.5 

# Intercepts
b0 = 1.5
d0 = 1
```

We know there is variablity in brain volume between subjects, so our actual brain volume for controls is:
    
###$BV_{hc} = \alpha_a * A + N(BV_0,\sigma_b)$

Where $\sigma_b$ is the standard deviation in brain volume between subjects, and N represents a normal distribution centered at $BV_0$

For patients, there is also variability in the disease metric, so the actual brain volume for patients is:

###$BV_{ms} = \alpha_a * A + N(BV_0,\sigma_b) + \alpha_d * D + N(D_0,\sigma_d)$

Where $\sigma_d$ represents the standard deviation in disease metric for patients.

We define the deviations for brain volume, $\sigma_b$ and disease metric, $\sigma_d$ **for each site**:


```python
#Brain Volume
sig_b = [0.08,0.08]

#Disease
sig_d = [0.2,0.2]
```

In this case, the populations are the same at each site, so the standard deviations are the same.

Now we must account for the scanner measurement. The scanner applies a linear transformation on the actual brain volume measurement:

###$$\hat{BV} = \alpha * BV + N(\gamma,\sigma_s)$$

* $\hat{BV}$ is the measure brain volume
* $\gamma$ is the measurement offset
* $\sigma_s$ is the variability of the scanner
* $\alpha$ is the scaling of the scanners

Finally, we need to define the linear transformation between sites:

###$\hat{BV_1} = \beta * \hat{BV_2} + S_{off}$

Where:

* $\hat{BV_1}$ is the measure brain volumes at site 1
* $\hat{BV_2}$ is the measured brain volumes at site 2
* $\beta$ is the scaling between the scanners
* $S_{off}$ is the offest between site 1 and site 2

For each site, we define $\sigma_s$, $\alpha$ and $\gamma$

```python
#Site
sig_s = [0.02,0.025] 
alpha = [0.9,1.1]
gamma = [0.05,-0.05]
```

####Creating datasets

For site 1:

```python
dis_1 = np.random.normal(d0,sig_d[0],n_ms) 
BV_hc_1 = alpha_a*age_hc + np.random.normal(b0,sig_b[0],n_hc)
BV_ms_1 = alpha_a*age_ms + np.random.normal(b0,sig_b[0],n_ms) + alpha_d*dis_1 + d0
```

```python
plot_corr([dis_1,age_ms],BV_ms_1,["DM","age"],"BV","Site 1")
```

For site 2:

```python
dis_2 = np.random.normal(d0,sig_d[0],n_ms) 
BV_hc_2 = alpha_a*age_hc + np.random.normal(b0,sig_b[0],n_hc)
BV_ms_2 = alpha_a*age_ms + np.random.normal(b0,sig_b[0],n_ms) + alpha_d*dis_2 + d0
```

```python
plot_corr([dis_2,age_ms],BV_ms_2,["DM","Age"],"BV","Site 2")
```

### Simulated Models

####Site 1: Patients

```python
s1_ms = run_model([age_ms,dis_1],BV_ms_1,sm.RLM)
```

####Site 2: Controls

```python
s2_hc = run_model([age_hc],BV_hc_2)
```

#### Site 2: Patients

```python
s2_ms = run_model([age_ms,dis_2],BV_ms_2,sm.RLM)
```

```python
def plot_coef_split(coef_list,err_list,coef_name_list,coef_site_list,title=""):
    N = len(coef_list) # num sites
    M = len(coef_list[0]) # num coefs
    
    fig, ax = subplots(1,M,figsize=(3*M,4)) #One plot per coef
    width = 1./(N)
    for j in range(M): #for each coef
        a = ax[j]
        a.set_xlim(.2,1.2)
        for i in range(N): #for each site
            X = [i*width] #there is one coef
            
            ppl.bar(a,X,[coef_list[i][j]],
                    yerr = [err_list[i][j]],
                    label=coef_site_list[i],
                    width=width,color=ppl.set2[i],annotate=True)
    
        a.set_xticks([])
        a.set_xticklabels([])
        a.set_title(coef_name_list[j])
    
    ppl.legend(a)
```

```python
plot_coef_split([s1_ms.params,s2_ms.params],
                [s1_ms.bse,s2_ms.bse],
                ["const","age","disease"],
                ["Site 1","Site 2"])
```

### Combining Sites

#### 1. Combine both sites

Controls

```python
age_all_hc = np.hstack((age_hc,age_hc))
BV_all_hc = np.hstack((BV_hc_1,BV_hc_2))
```

```python
combine_v1_hc = run_model([age_all_hc],BV_all_hc)
```

```python
plot_corr([age_all_hc],BV_all_hc,["age"],"BV","Controls: Combined Sites")
```

Patients

```python
BV_all_ms = np.hstack((BV_ms_1,BV_ms_2))
dis_all = np.hstack((dis_1,dis_2))
age_all_ms = np.hstack((age_ms,age_ms))
```

```python
combine_v1_ms = run_model([age_all_ms,dis_all],BV_all_ms,sm.RLM)
```

#### 2. Combine site by adding Protocol regressor

```python
protocol_ms = np.hstack((np.ones(n_ms),np.zeros(n_ms)))
```

```python
combine_v2_ms = run_model([age_all_ms,dis_all,protocol_ms],BV_all_ms,sm.RLM)
```

#### 3. Calibrate and then combine

```python
BV_ms_1_calib = calib.params[1] * BV_ms_1 + calib.params[0]
plot_corr([BV_ms_1_calib],BV_ms_2,["$\hat{BV_{1,calib}}$"],'$\hat{BV_2}$',"Calibrated Data")
```

```python
BV_all_calib = np.hstack((BV_ms_1_calib,BV_ms_2))
combine_v3 = run_model([age_all_ms,dis_all],BV_all_calib,sm.RLM)
```

```python
plot_coef_split([combine_v1_ms.params, combine_v2_ms.params[:3], combine_v3.params],
          [combine_v1_ms.bse, combine_v2_ms.bse[:3], combine_v3.bse],
          ["const","age","disease"],
          ["Combined","Protocol", "Calibrated"])
```
